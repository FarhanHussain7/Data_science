```python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset("tips")
df.head()
```

```python
# Distribution plot : It is used to show the distribution of a variable 
#  In simple words it is used to show the frequency of a variable 

sns.displot(df['total_bill'])
plt.show()
```

```python
df['size'].unique()
```

```python
# histplot :  it is used to plot the histogram of a variable 
# distplot : it is used to plot the distribution of a variable , it means it is used to plot the histogram of a variable
sns.histplot(df['total_bill'],bins=20,kde=True )
```

```python
#  subplots : it is used to create multiple plots in a single figure
# In simple words it is used to create multiple plots in a single figure

plt.subplot(2,2,1)
plt.subplot(2,2,2)
plt.subplot(2,2,3)
plt.subplot(2,2,4)
```

```python
# Jointplot : it is used to plot a scatter plot and a histogram in the same plot
# 
sns.jointplot(x="total_bill", y="tip", data=df, kind= "scatter")
```

```python
sns.jointplot(x= 'total_bill', y='tip', data=df,kind='hex')
sns.jointplot(x= 'total_bill', y='tip', data=df,kind='hex').savefig('jointplot.png')    
```

```python
sns.pairplot(df,hue="size",palette="husl")
sns.pairplot(df,aspect=0.5,palette="Set2")
sns.pairplot(df,corner=True,palette="Set3")
sns.pairplot(df,data=df,palette="Set1")
sns.pairplot(df,hue_order=["small","medium","large"],palette="Set1")
sns.pairplot(df,hue="size",palette="husl",markers=["o","s","D"])
```

```python
sns.rugplot(df)
sns.rugplot(df['total_bill'])
sns.rugplot(df['tip'])
```

```python
# Categorical Data Plots 
# It is used to represent the data in the form of categories
# There types 
# boxplot
# violin plot
# strip plot
# swarm plot
# bar plot
# count plot
# boxen plot


sns.load_dataset("tips")
```

```python
sns.countplot( data = df, x = 'sex')
sns.countplot( data = df, y = 'sex') 
sns.countplot(x= df['sex'],hue= df['smoker'])
```

```python
# All types of barplot
# sns.barplot(x= df['sex'], y= df['age'])
import numpy as np
sns.barplot(x= df['sex'], y= df['tip'], hue= df['smoker'])
sns.barplot(x= df['sex'], y= df['tip'], hue= df['smoker'])
sns.barplot(x= df['sex'], y= df['tip'], estimator= np.sum)
sns.barplot(x= df['sex'], y= df['tip'], estimator= np.mean)
sns.barplot(x= df['sex'], y= df['tip'], estimator= np.mean).set_title('Barplot of sex and tip')
```

```python
# bxplot
sns.boxplot(x="tip", y="day", data=df, palette="rainbow")
```
