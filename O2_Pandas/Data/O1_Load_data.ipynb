```python
# =========== Operation 2  - Extract Episode Information =============
# Extract episode information from the 'Title' column
def extract_episode(txt): # Function to extract episode info from title
    check = False # Flag to indicate if we are inside parentheses
    data = ""   # Variable to store extracted data
    for i in txt: # Loop through each character in the title
        if i == ")": # If we reach closing parenthesis
            check = False # Turn off the flag
            data += i  # Add closing parenthesis to data 
            return data  # Return the extracted data
        if i == "(": # If we reach opening parenthesis
            check = True # Turn on the flag
        if check == True: # If we are inside parentheses
            data += i # Add character to data
    return data # Apply the function to the 'Title' column
df['Title'].apply(extract_episode) # Apply the function to the 'Title' column
df['Episode_Info'] = df['Title'].apply(extract_episode) # Create a new column with extracted episode info
df['Episode_Info'] = df['Episode_Info'].str.replace("(", "").str.replace(")", "").str.replace(" eps", "") # Clean the new column by removing parentheses
df
df['Episode_Info'].value_counts() # Count occurrences of each unique value in the new column
df["Episode_Info"].astype(int) # Convert the new column to integer type
df["Episode_Info"] = pd.to_numeric(df["Episode_Info"], errors='coerce') # Convert to numeric, setting errors to NaN
```

```python
# ============= Operation 3 - Create new columns for time stamps =============
def extraction_time(txt): # Function to extract time information from title
    check = False # Flag to indicate if we are inside brackets
    data = ""   # Variable to store extracted data
    for i in range(len(txt)): # Loop through each character in the title
        if txt[i] == ")": # If we reach closing bracket
            for j in range(i+1,i+20): # Loop to get next 20 characters
                data += txt[j] # Add character to data
    return data # Apply the function to the 'Title' column
df['Title'].apply(extraction_time) # Apply the function to the 'Title' column
df['Total-time'] = df['Title'].apply(extraction_time) # Create a new column with extracted time info

```

```python
# =========== Operation 4 - Extract total months from time stamp =============
# Apr 2009 - Jul 2010 find months from this type of data
from dateutil.relativedelta import relativedelta
from datetime import datetime

def extract_months(txt): # Function to extract total months from time stamp
    try:
        start_str, end_str = txt.split(" - ") # Split the time stamp into start and end dates
        start_date = datetime.strptime(start_str.strip(), "%b %Y") # Convert start date string to datetime object
        end_date = datetime.strptime(end_str.strip(), "%b %Y") # Convert end date
        r = relativedelta(end_date, start_date) # Calculate the difference between end and start dates
        return r.years * 12 + r.months # Return total months
    except:
        return np.nan # Return NaN if there is an error

df['Total-time'].apply(extract_months) # Apply the function to the 'Total-time' column
df['Total-months'] = df['Total-time'].apply(extract_months) # Create a new column with total months
df 
```
