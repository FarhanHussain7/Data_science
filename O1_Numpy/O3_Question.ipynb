```python
# Valid sudoku board
board = [
  ["5","3",".",".","7",".",".",".","."],    
  ["6",".",".","1","9","5",".",".","."],
  [".","9","8",".",".",".",".","6","."],
  ["8",".",".",".","6",".",".",".","3"],
  ["4",".",".","8",".","3",".",".","1"],
  ["7",".",".",".","2",".",".",".","6"],
  [".","6",".",".",".",".","2","8","."],
  [".",".",".","4","1","9",".",".","5"],
  [".",".",".",".","8",".",".","7","9"] 
]

np.array(board)

# Check sudoku is valid or not
def isValidSudoku(board):
    rows = {}
    cols = {}
    boxes = {}
    
    for r in range(9):
        for c in range(9):
            val = board[r][c]
            if val == '.':
                continue
            
            # Check row
            if (r, val) in rows:
                return False
            rows[(r, val)] = True
            
            # Check column
            if (c, val) in cols:
                return False
            cols[(c, val)] = True
            
            # Check box
            box_index = (r // 3, c // 3)
            if (box_index, val) in boxes:
                return False
            boxes[(box_index, val)] = True
            
    return True

# 
isValidSudoku(board)

# Create a proper sudoku board
sudoku_board = np.array([
    [5,3,4,6,7,8,9,1,2],
    [6,7,2,1,9,5,3,4,8],
    [1,9,8,3,4,2,5,6,7],

    [8,5,9,7,6,1,4,2,3],
    [4,2,6,8,5,3,7,9,1],
    [7,1,3,9,2,4,8,5,6],

    [9,6,1,5,3,7,2,8,4],
    [2,8,7,4,1,9,6,3,5],
    [3,4,5,2,8,6,1,7,9]
    ])
# If the sum of each row, column and 3x3 sub-box is 45, then it's a valid sudoku
b = np.sum(sudoku_board, axis=1)
# b[b == 45]
for i in b:
    if i != 45:
        print("Invalid Sudoku")
        break
else:
    print("Valid Sudoku")

# slicing
sudoku_board[0:3, 0:3]
for i in range(0,9,3):
    for j in range(0,9,3):
        print(sudoku_board[i:i+3, j:j+3])   
```

```python
# Check if the sums of all rows, columns, and boxes are 45
row_sums = np.sum(sudoku_board, axis=1)
row_sums
col_sums = np.sum(sudoku_board, axis=0)
col_sums
box_sums = []
for i in range(0,9,3):
    for j in range(0,9,3):
        box = sudoku_board[i:i+3, j:j+3]
        box_sum = np.sum(box)
        box_sums.append(box_sum)    
box_sums
box_sums
all_sums = np.concatenate((row_sums, col_sums, box_sums))
all_sums
all_sums == 45
np.all(all_sums == 45)
```
