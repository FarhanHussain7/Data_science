```python
# Matplotlib :- It is a plotting library for creating static, animated, and interactive visualizations in Python.
# It is used to plot and visualize data in Python.
# It is built on top of NumPy and Pandas.
# It is widely used in data science and machine learning.
# Topics of Matplotlib
# 1. Basic of Matplotlib
# 2. Plotting Data
# 3. Customizing Plots
# 4. Subplots
# 5. Advanced Plotting



# Basic of Matplotlib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Matplotlib Basics
x = np.linspace(0,5,11) # Generate 11 evenly spaced points from 0 to 5
x 
y = x **2
y
```

```python
plt.plot(x,y) # plot the data
# plt.show() # show method is used to display the plot
plt.xlabel("X-axis") # set the label for x-axis
plt.ylabel("Y-axis") # set the label for y-axis
plt.title("Title") # set the title for the plot
plt.show() # display the plot
```

```python
# Subplots in Matplotlib
# Subplots are a grid of separate plots on the same figure. 
# You can create subplots using plt.subplots() and specify the number of rows and columns.
from matplotlib.pyplot import plot

plt.subplot(2,2,1) # 2 rows, 2 columns, 1st subplot
plt.plot(x,y)
plt.subplot(2,2,2) # 2 rows, 2 columns, 2nd subplot
plt.plot(x,y)
plt.subplot(2,2,3) # 2 rows, 2 columns, 3rd subplot
plt.plot(x**2,x**2)
plt.subplot(2,2,4) # 2 rows, 2 columns, 4th subplot
plt.plot(x,y)

```

```python
# Object Oriented Method :- It gives more control over the figure and axes
# It is more appropriate for more complex plots
# 
# fig is the entire figure or canvas
# axes is an array of subplot objects
fig, axes = plt.subplots(2,2)
print(axes)
# 
fig,plt.figure()
axis1 = fig.add_axes([0.1,0.1,0.8,0.8])
axis2 = fig.add_axes([0.2,0.5,0.4,0.3])
axis1.plot(x,y)
axis2.plot(x,y)
plt.show()
```

```python
# Desing in matplotlib
# It will add color to the chart
#  It will be a line chart
# It will be a bar chart
# It will be a histogram chart
# It will be a pie chart
# It will be a scatter chart
#  There are multiple ways to design the chart


#  create a project for designing 
import matplotlib.pyplot as plt
import numpy as np

# line chart
x = np.linspace(0, 2*np.pi, 400)
y = np.sin(x**2)
plt.plot(x, y, color='red', label='line chart')

# bar chart
x = np.arange(5)
y = np.random.rand(5)
plt.bar(x, y, color='green', label='bar chart')

# histogram
data = np.random.randn(1000)
plt.hist(data, bins=20, color='blue', label='histogram')

# pie chart
sizes = [15, 30, 45, 10]
labels = ['a', 'b', 'c', 'd']
plt.pie(sizes, labels=labels, colors=['yellow', 'orange', 'red', 'gray'],
        autopct='%1.1f%%', shadow=True, startangle=90, labeldistance=1.1)

# scatter chart
x = np.random.randn(50)
y = np.random.randn(50)
plt.scatter(x, y, color='purple', label='scatter chart')

plt.legend()
plt.show()

# Note:
# 1. plt.plot() for line chart
# 2. plt.bar() for bar chart
# 3. plt.hist() for histogram
# 4. plt.pie() for pie chart
# 5. plt.scatter() for scatter chart
 
```
